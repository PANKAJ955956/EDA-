```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
```

```python
print(df.head(10))
print(df.tail(10))
```

```python
print("Shape of dataset:", df.shape)
```

```python
# 4) Print all columns name
print("Columns:", df.columns)
```

```python
# 5) Find the data type of all the columns
print(df.dtypes)
```

```python
# 6) Print information and summary
print(df.info())
print(df.describe())
```

```python
# 7) Count Survived and show on pie chart
survival_counts = df["Survived"].value_counts()
plt.pie(survival_counts, labels=["Not Survived", "Survived"], autopct="%1.1f%%")
plt.title("Survival Count")
plt.show()
```

```python
# 8) Find out how many female passengers had travelled in first class and show on pie chart
female_first_class = df[(df["Sex"] == "female") & (df["Pclass"] == 1)]
plt.pie([len(female_first_class), len(df) - len(female_first_class)], labels=["1st Class Female", "Others"], autopct="%1.1f%%")
plt.title("Female Passengers in First Class")
plt.show()
```

```python
# 9) Female passengers who survived and age < 30
young_female_survived = df[(df["Survived"] == 1) & (df["Sex"] == "female") & (df["Age"] < 30)]
plt.pie([len(young_female_survived), len(df) - len(young_female_survived)], labels=["Survived < 30", "Others"], autopct="%1.1f%%")
plt.show()
```

```python
# 10) Male passengers who survived and age > 40
older_male_survived = df[(df["Survived"] == 1) & (df["Sex"] == "male") & (df["Age"] > 40)]
plt.pie([len(older_male_survived), len(df) - len(older_male_survived)], labels=["Survived > 40 Male", "Others"], autopct="%1.1f%%")
plt.show()
```

```python
# 11) Show age with 20 bins
df["Age"].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# 12) Age frequency with survived and not survived (Histogram)
sns.histplot(df, x="Age", hue="Survived", multiple="stack", bins=20)
plt.title("Age Frequency by Survival")
plt.show()
```

```python
# 13) Show Bar graph for Survived with male, female, class
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=df)
plt.title("Survival by Gender & Class")
plt.show()
```

```python
 #14) How many passengers are travelled in different classes (Bar Graph)
sns.countplot(x="Pclass", data=df)
plt.title("Passengers by Class")
plt.show()
```

```python
# 15) Passengers survived with class-wise (Bar Graph)
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate by Class")
plt.show()
```

```python
# 16) Show Q13 & Q14 in subplot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=df, ax=axes[0])
sns.countplot(x="Pclass", data=df, ax=axes[1])
axes[0].set_title("Survival by Gender & Class")
axes[1].set_title("Passengers by Class")
plt.show()
```

```python
 #17) Bar graph for 3rd class male, 1st class female survival
df_subset = df[(df["Pclass"] == 3) & (df["Sex"] == "male") | (df["Pclass"] == 1) & (df["Sex"] == "female")]
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=df_subset)
plt.title("Survival of 3rd Class Males & 1st Class Females")
plt.show()
```

```python
# 18) How many 1st class females survived/not survived
first_class_female = df[df["Pclass"] == 1][df["Sex"] == "female"]
sns.countplot(x="Survived", data=first_class_female)
plt.title("1st Class Female Survival")
plt.show()
```

```python
# 19) Map Sex column (male=1, female=0)
df["Sex"] = df["Sex"].map({"male": 1, "female": 0})
print(df.head())
```

```python
# 20) Find Null values
print("Null values:\n", df.isnull().sum())
```

```python
# 21) Replace Null value with Median
df.fillna(df.median(), inplace=True)
```

```python
# 22) Drop unwanted columns
df.drop(columns=["Cabin", "Ticket", "Name"], inplace=True)
```

```python
print("Final cleaned dataset:\n", df.head())
```
